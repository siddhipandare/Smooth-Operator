```python
import matplotlib.pyplot as plt
```

```python
# Read lines from file exp_smoothing_data.txt

x = []
y = []
labels = []

with open('exp_smoothing_data.txt') as f:
    data = f.readlines()
    for line in data:
        line = line.split()
        labels.append(line[0])
        x.append(float(line[1]))
        y.append(float(line[2]))

# Plot a 2D scatter plot with x, y and labels
# Ensure your labels are not overlapping by making font size smaller

plt.scatter(x, y)
for i, txt in enumerate(labels):
    if i == 9:
        continue
    plt.annotate(txt, (x[i], y[i]), fontsize=8)

# Y axis label is Number of Votes
# X axis label is Seasonality Strength
plt.xlabel('Seasonality Strength')
plt.ylabel('Number of Votes')


# Make title as "Number of Votes vs Seasonality Strength metric"
plt.title('Number of Votes vs Seasonality Strength metric')

# Change the color of the points with label sine.csv, climate_jfk_tmax.json and monthly_beer_austria.csv to red
plt.scatter(x[1], y[1], color='red')
plt.scatter(x[2], y[2], color='red')
plt.scatter(x[3], y[3], color='red')



plt.show()
```
